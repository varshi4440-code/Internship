```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\varshitha v\Desktop\ds_ai_internship\src\day28\homeprices.csv")
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='blue', marker = 'o') 
```

```python
area = df.drop('price',axis = 'columns')
area
```

```python
price = df.price
price
```

```python
#equation for linear regression y = mx+c

reg = linear_model.LinearRegression()
reg.fit(area,price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#to predict the new area 3500
new_data =pd.DataFrame([[3500]],columns=['area'])
prediction = reg.predict(new_data)
prediction = np.round(prediction,2)
prediction
```

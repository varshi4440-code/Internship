```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df =pd.read_csv(r"C:\Users\varshitha v\Desktop\ds_ai_internship\src\day29\insurance_data.csv")
df.head()
```

```python
df.shape
```

```python
plt.scatter(df.age,df.bought_insurance,color='red',marker='+')
```

```python
x=df[['age']]
y=df['bought_insurance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8, random_state=10)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
df['probs']=model.predict_proba(df[['age']])[:,1]
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+', color='red')
plt.plot(df['age'],df['probs'],color='blue')
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.title("Logistic Regression signoid curve")
plt.show()
```

```python
df_sorted = df.sort_values('age')

plt.scatter(df['age'],df['bought_insurance'],marker='+',color='red')
plt.plot(df_sorted['age'],df_sorted['probs'],color='blue')
plt.show()
```

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def prediction_function(age):
    z=0.112*age+(-4.20)
    y= sigmoid(z)
    return y
```
